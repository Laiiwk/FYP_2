```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#data_path =  'output/extracted_error_features.csv'
df = pd.read_csv('../output/extracted_error_features.csv')
```

```python
# Get average error occurrence per month, grouped by error type
error_counts_by_month = df.groupby(['month', 'error_level_encoded']).size() \
                        .unstack(fill_value=0) \
                        .mean(axis=0)  # Average across months

# Plotting
plt.figure(figsize=(10, 6))  # Adjust figure size as needed
error_counts_by_month.plot(kind='bar')
plt.xlabel('Error Level (Encoded)')
plt.ylabel('Average Monthly Occurrence')
plt.title('Average Error Occurrence by Type and Month')
plt.legend(title='Month')  # Display a legend for the months
plt.show()
```

```python
# Calculate monthly averages for error_level_encoded (assuming it's numerical)
monthly_averages = df.groupby(df['month'])['error_level_encoded'].mean()
print(monthly_averages) 
```

```python
# Count occurrences of error types by month
error_counts_by_month = df.groupby([df['month'], 'error_level_encoded'])['error_level_encoded'].count()
print(error_counts_by_month)
```

```python
# precentage of error level by month
error_counts_by_month = error_counts_by_month.groupby(level=0).apply(lambda x: 100 * x / float(x.sum()))
#print(error_counts_by_month)
print (error_counts_by_month.unstack(fill_value=0))  # Display the table
```

```python
# Filter for the desired month
month = 4
df_filtered = df[df['month'] == month]

# Pivot error counts into a 'day' vs. 'error_level_encoded' matrix
error_counts_pivot = df_filtered.pivot_table(
    index='day', columns='error_level_encoded', values='pid', aggfunc='count'
).fillna(0).astype(int)  # Fill missing values with 0

# Heatmap visualization
plt.figure(figsize=(8, 6))
sns.heatmap(error_counts_pivot, cmap='viridis', annot=True, fmt='d')  # 'viridis' cmap is a good default
plt.xlabel('Error Level (Encoded)')
plt.ylabel('Day of Month')
plt.title(f'Error Occurrence Heatmap for Month {month}')
plt.show()
```

```python
error_level = 3
month = 3

# Filter for the desired month and error level
df_filtered = df[(df['month'] == month) & (df['error_level_encoded'] == error_level)]

error_counts_pivot = df_filtered.pivot_table(
        index='day', columns='error_level_encoded', values='pid', aggfunc='count'
    ).fillna(0)  

 # Select only the column of the desired error level for heatmap generation
heatmap_data = error_counts_pivot[error_level] 

# Heatmap visualization
plt.figure(figsize=(8, 6))
sns.heatmap(heatmap_data.to_frame(), cmap='viridis', annot=True, fmt='d')  
plt.xlabel('Day of Month') # Since we now have only one column
plt.ylabel('') 
plt.title(f'Error Level {error_level} Occurrence Heatmap for Month {month}')
plt.show()
```
